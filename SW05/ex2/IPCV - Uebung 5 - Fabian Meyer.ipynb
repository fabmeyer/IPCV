## 3: Laplace-Filter

```python
img = cv2.imread("../imdemos/peppers.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_original = img
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
img_blurred = img

# h channel from hsv image
h = img_blurred[:,:,2]

h_blur = cv2.blur(h, (3,3))

img_blurred[:,:,2] = h_blur

img = cv2.cvtColor(img, cv2.COLOR_HSV2RGB)
img_blurred = cv2.cvtColor(img_blurred, cv2.COLOR_HSV2RGB)

kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
img_blurred_laplace = cv2.filter2D(img_blurred, -1, kernel)

plt.subplot(1, 2, 1)
plt.imshow(img_original)

plt.subplot(1, 2, 2)
plt.imshow(img_blurred_laplace)
```

## 4: Betrag des Gradienten

```python
img = cv2.imread("../Zebra.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

sobelx = cv2.Sobel(img, cv2.CV_64F,1,0,ksize=5)  # x
sobely = cv2.Sobel(img, cv2.CV_64F,0,1,ksize=5)  # y
laplace = np.sqrt(np.add(np.square(sobelx), np.square(sobely))) # square root of sum of squares of derivatives

plt.subplot(2,2,1)
plt.imshow(img, cmap = 'gray')

plt.subplot(2,2,2)
plt.imshow(sobelx, cmap = 'gray')

plt.subplot(2,2,3)
plt.imshow(sobely, cmap = 'gray')

plt.subplot(2,2,4)
plt.imshow(laplace, cmap = 'gray')

plt.show()
```

## 5: Gauss- und Laplacepyramide

```python
img = cv2.imread("../imdemos/peppers.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

level0 = img
level1 = cv2.pyrDown(level0)
level2 = cv2.pyrDown(level1)
level3 = cv2.pyrDown(level2)

laplacian0 = cv2.subtract(level0, cv2.pyrUp(level1))
laplacian1 = cv2.subtract(level1, cv2.pyrUp(level2))
laplacian2 = cv2.subtract(level2, cv2.pyrUp(level3))

plt.subplot(1, 3, 1)
plt.imshow(img)

plt.subplot(1, 3, 2)
plt.imshow(level1)

plt.subplot(1, 3, 3)
plt.imshow(level2)

plt.show()

plt.subplot(1, 3, 1)
plt.imshow(4 * laplacian0)

plt.subplot(1, 3, 2)
plt.imshow(4 * laplacian1)

plt.subplot(1, 3, 3)
plt.imshow(4 * laplacian2)

plt.show()
```
